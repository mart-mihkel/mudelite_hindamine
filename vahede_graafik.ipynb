```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
q = np.linspace(1e-10, 1, 1000)
beta = np.array([0.5, 0.75, 0.9])
```

```python
var_N = [1 / (b * (2 * q - 1) ** 2) - 1 for b in beta]
var_N[0]
```

```python
for v_N in var_N:
    plt.plot(v_N, q)

plt.plot([0, 1e6], [0.5, 0.5], 'k--')
    
plt.title('igsuste lhendite vahe relatiivne viga')

plt.legend([f' = {b}' for b in beta])
plt.ylabel('q')
plt.xlabel('dispersioon * N')

plt.show()
```
