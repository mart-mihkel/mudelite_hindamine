```python
import numpy as np
import matplotlib.pyplot as plt
```

\begin{equation}
    \frac{1}{(2q-1)(p_A-p_B)} - 1
\end{equation}

```python
q = np.linspace(1e-10, 1, 100)
vahed = np.array([0.1, 0.25, 0.75, 0.9])
asjad = [1 / ((2 * q - 1) * v) - 1 for v in vahed]
```

```python
for asi in asjad:
    plt.plot(asi, q)

plt.plot([0, 1e4], [0.5, 0.5], 'k--')
    
plt.title('Õigsuste lähendite vahe relatiivne viga')

plt.legend([f'p_A - p_B = {v}' for v in vahed])
plt.ylabel('q')
plt.xlabel('dispersioon * N')

plt.show()
```

```python
for asi in asjad:
    plt.plot(asi, q)

plt.plot([0, 1e4], [0.5, 0.5], 'k--')

plt.title('Õigsuste lähendite vahe relatiivne viga')

plt.xscale('log')

plt.legend([f'p_A - p_B = {v}' for v in vahed])
plt.ylabel('q')
plt.xlabel('dispersioon * N')

plt.show()
```

\begin{equation}
    \frac{1}{\beta(2q-1)^2} - 1
\end{equation}

```python
q = np.linspace(1e-10, 1, 100)
beta = np.array([0.5, 0.75, 0.9])
asjad = [1 / (b * (2 * q - 1) ** 2) - 1 for b in beta]
```

```python
for asi in asjad:
    plt.plot(asi, q)

plt.plot([0, 1e4], [0.5, 0.5], 'k--')
    
plt.title('Õigsuste lähendite vahe relatiivne viga')

plt.legend([f'β = {b}' for b in beta])
plt.ylabel('q')
plt.xlabel('dispersioon * N')

plt.show()
```

```python
for asi in asjad:
    plt.plot(asi, q)

plt.plot([0, 1e4], [0.5, 0.5], 'k--')

plt.title('Õigsuste lähendite vahe relatiivne viga')

plt.xscale('log')

plt.legend([f'β = {b}' for b in beta])
plt.ylabel('q')
plt.xlabel('dispersioon * N')

plt.show()
```
